Question 5 Use the adapted parameters to obtain reproducible results.

In this case, we use 'random_state' to produce reproducible results in the program.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.3, random_state=42)
accuracies2 = []

for i in range(1, 32):
    if(i%2!=0):
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(x_train, y_train)
        y_pred = knn.predict(x_test)
        y_true=y_test
        accuracy = accuracy_score(y_test, y_pred)
        accuracies2.append(accuracy)
```

```python
print(accuracies2)
```

We infer that adding a 'random_state' makes the splitting consistent making the program reprodicable.
